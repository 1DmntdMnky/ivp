# Difference between opening an image using PIL and mimage
- Image is opening the image (PIL Library)
- imread will convert the image to matrix
- imshow will convert matrix to image on the screen



```python
import matplotlib.pyplot as plt
import matplotlib.image as mimage
image1 = mimage.imread('cat.png')
image1.shape
```

```python
imagearr= image1[100:110,50:60,0]
image2 = imagearr * 256 
plt.imshow(image2)
# plt.imshow(image1)

```

```python
image3=image1.copy()
image3[image3<0.8]=0 #Changing all the values of pixels less than 0.8 to 0 to darken the image.
print(image3)
plt.imshow(image3)
```

```python
from PIL import Image
import matplotlib.image as mimage
import matplotlib.pyplot as plt
peppers = Image.open('peppers.jpg')
plt.figure()
plt.subplot(2,1,1)
plt.imshow(peppers)
plt.title('png')
peppers.save('peppers.jpg')
peppers2 = Image.open('peppers.jpg')
plt.subplot(2,1,2)
plt.imshow(peppers2)
plt.title('jpg')
```

## What does subplot() do?
It changes the allocation of the images, so if we use subplot(1,2,1) and subplot (1,2,2) the two images will be displayed side by side horizontally. 

When we use subplot(2,1,1) and subplot(2,2,2), the image will be displayed vertically.


```python
from skimage.io import imread, imshow
from skimage import color
import matplotlib.pyplot as plt
image5=imread('ishihara.jpg')
image5_gray=color.rgb2gray(image5)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image5)
plt.subplot(1,2,2)
plt.imshow(image5_gray,cmap='gray')
```
