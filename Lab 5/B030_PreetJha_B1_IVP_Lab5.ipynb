```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct
import cv2
```

```python
img=cv2.imread('lenna_grey.jpg',0);
row,col=img.shape
N=8
```

```python
img_dct = np.zeros((row, col), dtype=int)
for r in range(0, row//N):
    for c in range(0, col//N):
        #img_dct[r*N:(r+1)*N, c*N:(c+1)*N] = 
        temp1= dct(img[r*N:(r+1)*N, c*N:(c+1)*N])
        temp1[1:7,1:7]=0
        img_dct[r*N:(r+1)*N, c*N:(c+1)*N]=temp1
        #img_dct=dct(img)
plt.imshow(img_dct, cmap='gray')
```

```python
img_idct=np.zeros((row,col), dtype=int)
for r in range(0, row//N):
    for c in range(0, col//N):
        img_idct[r*N:(r+1)*N, c*N:(c+1)*N]=idct(img_dct[r*N:(r+1)*N, c*N:(c+1)*N])

plt.imshow(img_idct,cmap='gray')
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("original")

plt.subplot(1,3,2)
plt.imshow(img_dct, cmap='gray')
plt.title("dct")

plt.subplot(1, 3, 3)
plt.imshow(img_idct, cmap='gray')
plt.title("idct")
```

# Conclusion
- DCT is performed on 'lenna_gray.jpg'.
- We will take the DCT of the original image is computed and the inverse of the same DCT is calculated, we get the same image back.
- If last 4 rows and last 4 cols of DCT of the original image are converted to 0, then inverse DCT of the given image shows almost similar quality to the original image.
- This shows that top 48 out of 64 coefficients can be retained to get acceptable quality of the image.
- If the last 6 rows and 6 cols are ignored to further compress the image, then the inverse DCT of the given image will show significant distortion.
- Same results are observed if the given image is 'test.png'.
